```python
import pandas as pd
import json
import numpy as np
import os
```

```python
releases = pd.DataFrame(columns={'project','release','release_date'})
```

```python
with open('releases.json') as f:
    lines = f.readlines()
```

```python
combined = ''
for i in lines:
    combined = combined+(i.strip('\n'))
```

```python
js = json.loads(combined)
```

```python
for i,j in js.items():
    for k,l in j.items():
        releases = releases.append({'project':i,'release':k,'release_date':l},ignore_index=True)
```

```python
releases = releases.sort_values(['project','release_date']).reset_index(drop=True)
```

```python
releases['release_date']=pd.to_datetime(releases['release_date'])
```

```python
releases['delta'] = releases.groupby('project')['release_date'].diff().fillna(pd.Timedelta(seconds=0))
```

```python
releases_gp = releases.groupby('project').agg({'release':'count','delta':'sum'}).reset_index()
releases_gp['av_delta'] = (releases_gp['delta']/releases_gp['release']) / np.timedelta64(1, 'D')
```

```python
repo = pd.read_csv('repo_data.csv')
```

```python
repo = repo.rename(columns={'Repo Name':'project'})
```

```python
repo = repo.sort_values(['project']).reset_index(drop=True)
```

```python
releases_gp = releases_gp.sort_values(['project']).reset_index(drop=True)
```

```python
repo=repo.merge(releases_gp, how='left',on='project')
```

```python
repo[~repo['release'].isna()].reset_index(drop=True).to_csv('repo_release_matched.csv')
```
